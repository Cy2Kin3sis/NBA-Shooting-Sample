# NBA Shooting Data
This dataset contains shooting statistics for four different players during the 2021 NBA Playoffs.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage

nba = pd.read_csv("nba_players_shooting.csv", index_col=0)
x = nba['X']
y = nba['Y']

unique_players = ['Seth Curry', 'Chris Paul', 'Russell Westbrook', 'Trae Young']

def shot_chart():
    marker_styles = {'MADE': 'o', 'MISSED': 'x'}
    color_s = ['#006BB6', '#E56020', '#002B5C', '#C1D32F']
    
    print("Select an option:")
    print("[0] Plot all players combined")
    print("[1] Plot Seth Curry")
    print("[2] Plot Chris Paul")
    print("[3] Plot Russell Westbrook")
    print("[4] Plot Trae Young")
    option = input("Enter option: ")
    option = int(option)
    count = -1
    
    halfcourt_img = plt.imread('court.png')
    
    plt.figure(figsize=(12, 11))

    plt.imshow(halfcourt_img, extent=[-25.6,25.6,0,47])

    if option == 1:
        player_data = nba[nba['SHOOTER'] == unique_players[option - 1]]
        plt.scatter(player_data[player_data['SCORE'] == "MISSED"]['X'], 
                player_data[player_data['SCORE'] == "MISSED"]['Y'], 
                color=color_s[option - 1], 
                marker=marker_styles['MISSED'], 
                label=f'{unique_players[option - 1]} - Missed Shots')
        plt.scatter(player_data[player_data['SCORE'] == "MADE"]['X'], 
                player_data[player_data['SCORE'] == "MADE"]['Y'], 
                color=color_s[option - 1],
                marker=marker_styles['MADE'], 
                label=f'{unique_players[option - 1]} - Made Shots')
        plt.title("2021 Season - Seth Curry Shot Chart")

    elif option == 2:
        player_data = nba[nba['SHOOTER'] == unique_players[option - 1]]
        plt.scatter(player_data[player_data['SCORE'] == "MISSED"]['X'], 
                player_data[player_data['SCORE'] == "MISSED"]['Y'], 
                color=color_s[option - 1],
                marker=marker_styles['MISSED'], 
                label=f'{unique_players[option - 1]} - Missed Shots')
        plt.scatter(player_data[player_data['SCORE'] == "MADE"]['X'], 
                player_data[player_data['SCORE'] == "MADE"]['Y'], 
                color=color_s[option - 1],
                marker=marker_styles['MADE'], 
                label=f'{unique_players[option - 1]} - Made Shots')
        plt.title("2021 Season - Chris Paul Shot Chart")

    elif option == 3:
        player_data = nba[nba['SHOOTER'] == unique_players[option - 1]]
        plt.scatter(player_data[player_data['SCORE'] == "MISSED"]['X'], 
                player_data[player_data['SCORE'] == "MISSED"]['Y'], 
                color=color_s[option - 1], 
                marker=marker_styles['MISSED'], 
                label=f'{unique_players[option - 1]} - Missed Shots')
        plt.scatter(player_data[player_data['SCORE'] == "MADE"]['X'], 
                player_data[player_data['SCORE'] == "MADE"]['Y'], 
                color=color_s[option - 1], 
                marker=marker_styles['MADE'], 
                label=f'{unique_players[option - 1]} - Made Shots')
        plt.title("2021 Season - Russell Westbrook Shot Chart")
    
    elif option == 4:
        player_data = nba[nba['SHOOTER'] == unique_players[option - 1]]
        plt.scatter(player_data[player_data['SCORE'] == "MISSED"]['X'], 
                player_data[player_data['SCORE'] == "MISSED"]['Y'], 
                color=color_s[option - 1], 
                marker=marker_styles['MISSED'], 
                label=f'{unique_players[option - 1]} - Missed Shots')
        plt.scatter(player_data[player_data['SCORE'] == "MADE"]['X'], 
                player_data[player_data['SCORE'] == "MADE"]['Y'], 
                color=color_s[option - 1], 
                marker=marker_styles['MADE'], 
                label=f'{unique_players[option - 1]} - Made Shots')
        plt.title("2021 Season - Trae Young Shot Chart")

    elif option == 0:
        for player in unique_players:
            count = count + 1
            player_data = nba[nba['SHOOTER'] == player]
            plt.scatter(player_data[player_data['SCORE'] == "MISSED"]['X'], 
                    player_data[player_data['SCORE'] == "MISSED"]['Y'], 
                    color=color_s[count], 
                    marker=marker_styles['MISSED'], 
                    label=f'{player} - Missed Shots')
            plt.scatter(player_data[player_data['SCORE'] == "MADE"]['X'], 
                    player_data[player_data['SCORE'] == "MADE"]['Y'], 
                    color=color_s[count], 
                    marker=marker_styles['MADE'], 
                    label=f'{player} - Made Shots')

    else:
        plt.title("No graph available.")
        print("Graph empty.")


    plt.xlabel("X Coordinate")
    plt.ylabel("Y Coordinate")
    plt.legend()
    plt.show()

def shot_range_distribution():
    seth = nba[(nba['SHOOTER'] == 'Seth Curry') & (nba['SCORE'] == 'MADE')]
    chpl = nba[(nba['SHOOTER'] == 'Chris Paul') & (nba['SCORE'] == 'MADE')]
    russ = nba[(nba['SHOOTER'] == 'Russell Westbrook') & (nba['SCORE'] == 'MADE')]
    trae = nba[(nba['SHOOTER'] == 'Trae Young') & (nba['SCORE'] == 'MADE')]

    seth_shot = seth['RANGE'].value_counts()
    chpl_shot = chpl['RANGE'].value_counts()
    russ_shot = russ['RANGE'].value_counts()
    trae_shot = trae['RANGE'].value_counts()

    fig, axs = plt.subplots(2, 2)

    axs[0, 0].pie(seth_shot, labels=seth_shot.index, autopct='%1.1f%%')
    axs[0, 0].set_title("Seth Curry")

    axs[0, 1].pie(chpl_shot, labels=seth_shot.index, autopct='%1.1f%%')
    axs[0, 1].set_title("Chris Paul")

    axs[1, 0].pie(russ_shot, labels=seth_shot.index, autopct='%1.1f%%')
    axs[1, 0].set_title("Russell Westbrook")

    axs[1, 1].pie(trae_shot, labels=seth_shot.index, autopct='%1.1f%%')
    axs[1, 1].set_title("Trae Young")


    fig.suptitle("Made Shots Distribution by select players, 2021 playoffs")
    plt.tight_layout()
    plt.show()
    
def defended_by():
    def_table = nba.pivot_table(index='SHOOTER', columns='DEFENDER', values='SCORE', aggfunc=lambda x: f"{x[x=='MADE'].count()}/{x.count()}", fill_value='---', margins=True, margins_name='Total')

    print("Shooting stats")
    print(def_table)

    pivot_table_filtered = def_table.drop('Total', errors='ignore')

    shooting_percentages = pivot_table_filtered.applymap(lambda x: float(x.split('/')[0]) / float(x.split('/')[1]) if x != '---' else None)

    shooting_percentages_filtered = shooting_percentages[shooting_percentages != '---'].astype(float)

    shooting_percentages_filtered.plot(kind='bar', width=0.8, figsize=(10, 6))
    plt.title('Shooter\'s Shooting Percentage Against Each Defender')
    plt.xlabel('Shooter')
    plt.ylabel('Shooting Percentage')
    plt.xticks(rotation=0)
    plt.legend(title='Defended By')
    plt.grid(axis='y')
    plt.show()

    
while True:
    print("NBA Shooting Data Menu")
    print("Choose which visualization you want to see:")
    print("[1] Shot Charts")
    print("[2] Shot Distributions")
    print("[3] Shooter vs. Defender")
    print("Press any key to exit")
    choice = input("Enter your choice: ")
    
    if(choice == '1'):
        shot_chart()
    elif(choice == '2'):
        shot_range_distribution()
    elif(choice == '3'):
        defended_by()
    else:
        print("Thank you for using this program!")
        break
    
```

## Data Dictionary
|variable |class     |description                                                  |
|:--------|:---------|:------------------------------------------------------------|
|SHOOTER  |String    |Name of the player taking the shot                           |
|X        |float     |Horizontal distance of the shot taken from the basket in ft  |
|Y        |float     |Vertical distance of the shot taken from the basket in ft    |
|RANGE    |String    |Radius range of the shot taken from the basket in ft         |
|DEFENDER |String    |Name of the player defending the shot                        |
|SCORE    |String    |'MADE' if shot is scored, else 'MISSED'                      |

[Source](https://www.nba.com/stats/players/shooting/?Season=2020-21&SeasonType=Playoffs&PerMode=Totals) of dataset.
